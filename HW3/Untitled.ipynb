```python
import urllib.request as req

url="https://www.ptt.cc/bbs/LoL/index.html"

#建立物件，附加Request Headers
request = req.Request(url, headers={
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
})
with req.urlopen(request) as response:
    data=response.read().decode("utf-8")
print(data) 
```

```python
#解析原始碼
import bs4
root=bs4.BeautifulSoup(data, "html.parser")
print(root.title)
```

```python
#title文字
print(root.title.string)
```

```python
#尋找標籤
titles=root.find("div",class_="title")
print(titles)
print(titles.a.string)
```

```python
#找尋所有的標籤
title_all=root.find_all("div",class_="title")
print(title_all)
```

```python
for title in title_all:
    if title.a !=None:
        print(title.a.string)
```

```python
import requests
from bs4 import BeautifulSoup
import json

# 發送HTTP GET請求，獲取網頁內容
url = "https://www.ptt.cc/bbs/LoL/index.html"
response = requests.get(url)

# 解析HTML代碼，提取需要的數據
soup = BeautifulSoup(response.text, 'html.parser')
posts = []
for article in soup.find_all('article'):
    title = article.h2.text.strip()
    excerpt = article.find(class_='sc-1v1d5rx-3').text.strip()
    href = article.h2.a.get('href')
    post = {
        'title': title,
        'excerpt': excerpt,
        'href': href
    }
    posts.append(post)

# 將數據以JSON格式儲存到檔案中
with open('lol.json', 'w', encoding='utf-8') as f:
    json.dump(posts, f, ensure_ascii=False, indent=4)
```

```python
import pandas as pd

tailand = pd.read_csv("tailand.csv")
tailand
```

```python

# 找到status_types欄位，並從中選取video和photo的資料列
video_rows = tailand[tailand['status_type'] == 'video']
photo_rows = tailand[tailand['status_type'] == 'photo']

# 使用value_counts()方法分別統計video和photo的數量
num_videos = video_rows['status_type'].value_counts()[0]
num_photos = photo_rows['status_type'].value_counts()[0]

print(f"video的數量為：{num_videos}")
print(f"photo的數量為：{num_photos}")
```

```python
# 找到出現次數最多分享的貼文
most_shares = tailand.sort_values(by='num_shares', ascending=False).iloc[0]

# 打印出現次數最多的留言和出現次數
print(most_shares)
```

```python
# 根據互動量進行排序，獲取互動最多的一次
most_interactive = tailand.sort_values(by='num_reactions', ascending=False).iloc[0]

# 打印該次互動的詳細資訊
print(most_interactive)
```

```python

# 計算每種類型出現的次數
type_counts = tailand['status_type'].value_counts()

# 找到出現次數最多的類型
most_common_type = type_counts.index[0]

# 打印出現次數最多的類型
print('出現次數最多的類型是:', most_common_type)
```

```python
# 找到出現次數最多的留言
most_comments = tailand.sort_values(by='num_comments', ascending=False).iloc[0]

# 打印出現次數最多的留言和出現次數
print(most_comments)
```

```python

# 確認互動量欄位名稱，例如按讚、分享、評論
interactions = ['num_reactions', 'num_comments', 'num_shares']

# 將每篇貼文的互動量相加
total_interactions = tailand[interactions].sum(axis=1)

# 統計貼文總數
status_id = len(tailand)

# 將總互動量除以貼文總數，得到平均互動量
avg_interactions = total_interactions.mean()

print(f"所有貼文的平均互動量為：{avg_interactions}")
```

```python
# 找到likes、love、sad和angry欄位，並計算它們的總和
total_reactions = tailand['num_likes'].sum() + tailand['num_loves'].sum() + tailand['num_sads'].sum() + tailand['num_angrys'].sum()
+ tailand['num_wows'].sum() + tailand['num_hahas'].sum()
# 分別計算likes、love、sad和angry的數量佔總和的百分比
percent_likes = (tailand['num_likes'].sum() / total_reactions) * 100
percent_love = (tailand['num_loves'].sum() / total_reactions) * 100
percent_sad = (tailand['num_sads'].sum() / total_reactions) * 100
percent_angry = (tailand['num_angrys'].sum() / total_reactions) * 100
percent_wow = (tailand['num_wows'].sum() / total_reactions) * 100
percent_haha = (tailand['num_hahas'].sum() / total_reactions) * 100

print(f"likes佔總和的百分比：{percent_likes}%")
print(f"love佔總和的百分比：{percent_love}%")
print(f"sad佔總和的百分比：{percent_sad}%")
print(f"angry佔總和的百分比：{percent_angry}%")
print(f"wows佔總和的百分比：{percent_wow}%")
print(f"hahas佔總和的百分比：{percent_haha}%")
```

```python
# 計算總共的reactions、comments和shares數量
total_reactions = tailand['num_comments'].sum() + tailand['num_shares'].sum() + tailand['num_reactions'].sum()

# 計算reactions、comments和shares數量分別佔總合的百分比
percent_comments = (tailand['num_comments'].sum() / total_reactions) * 100
percent_shares = (tailand['num_shares'].sum() / total_reactions) * 100
percent_reactions = (tailand['num_reactions'].sum() / total_reactions) * 100

print(f"reactions佔總和的百分比：{percent_reactions:.2f}%")
print(f"comments佔總和的百分比：{percent_comments:.2f}%")
print(f"shares佔總和的百分比：{percent_shares:.2f}%")
```

```python
# 找到 loves 最多的貼文
max_loves_index = tailand['num_loves'].idxmax()

# 輸出貼文的內容和 loves 數量
print("The post with the most num_loves is:")
print(tailand.loc[max_loves_index]['status_id'])
print("Number of num_loves:", tailand.loc[max_loves_index]['num_loves'])
```

```python
# 找到 wows 最多的貼文
max_wows_index = tailand['num_wows'].idxmax()

# 輸出貼文的內容和 wows 數量
print("The post with the most num_wows is:")
print(tailand.loc[max_wows_index]['status_id'])
print("Number of num_wows:", tailand.loc[max_wows_index]['num_wows'])
```
